```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(0, math.pi*2, 0.05)

y1 = np.sin(x)
y2 = np.cos(x)
y3= np.tan(x)
y4 = np.tanh(x)

fig, axis = plt.subplots(2,2)

axis[0, 0].plot(x, y1)
axis[0, 0].set_title("Sine Function")

axis[0,1].plot(x,y2)
axis[0,1].set_title('cos')

```
